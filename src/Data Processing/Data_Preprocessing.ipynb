## **checking for Outliers**

```python
plt.figure(figsize=(15,10),canvas="white")
plotnumber=1
for column in num_data:
  if plotnumber<=10:
    ax=plt.subplot(4,3,plotnumber)
    sns.boxplot(x=data[column])
    plt.xlabel(column,fontsize=10)
    plt.ylabel("Performence Rate",fontsize=10)
  plotnumber+=1
plt.tight_layout()
```

### **Handle Outliers for NumCompaniesWorked**

```python
sns.histplot(data=data,x=data["NumCompaniesWorked"],kde=True)
```

```python
#sns.boxplot(x=data.NumCompaniesWorked)
```

```python
Q3=data["NumCompaniesWorked"].quantile(0.75)
print("upper_quartile",Q3)
Q1=data["NumCompaniesWorked"].quantile(0.25)
print("lower_quartile",Q1)
```

```python
IQR=Q3-Q1
IQR
```

```python
upper_limit=Q3+1.5*IQR
print("upperlimit",upper_limit)
lower_limit=Q1-1.5*IQR
print("lowerlimit",lower_limit)
```

```python
data.loc[data["NumCompaniesWorked"]>upper_limit]
```

```python
len(data.loc[data["NumCompaniesWorked"]>upper_limit])/len(data)
```

```python
data.loc[data["NumCompaniesWorked"]>upper_limit,"NumCompaniesWorked"]=data["NumCompaniesWorked"].median()
```

```python
sns.boxplot(x=data.NumCompaniesWorked)
```

```python
plt.figure(figsize=(15,10),canvas="white")
plotnumber=1
for column in num_data:
  if plotnumber<=10:
    ax=plt.subplot(4,3,plotnumber)
    sns.boxplot(x=data[column])
    plt.xlabel(column,fontsize=10)
    plt.ylabel("Performence Rate",fontsize=10)
  plotnumber+=1
plt.tight_layout()
```

**Sucessfully Removed Outliers**
