### **Find Correlation:**

```python
data.corr()
```

```python
plt.figure(figsize=(12, 10))
correlation_matrix = data.drop("Performance_Rating", axis=1).corr()
heat = sns.heatmap(correlation_matrix, cmap="viridis", annot=True, fmt=".2f")
plt.title("HeatMap Analysis", fontsize=16)
heat.set_xticklabels(heat.get_xticklabels(), fontsize=5)
heat.set_yticklabels(heat.get_yticklabels(), fontsize=5)
plt.xlabel("Columns", fontsize=8)
plt.ylabel("Columns", fontsize=8)
plt.show()
```

```python
def correlation(dataset,threshold):
  col_corr=set()
  matrix=dataset.corr()
  for i in range(len(matrix.columns)):
    for j in range(i):
      if(matrix.iloc[i,j]>threshold):
        col_name=matrix.columns[i]
        col_corr.add(col_name)
      return col_corr
```

```python
features=correlation(data,0.4)
len(set(features))
```

# **Feature Engineering Report:**


**Feature Encoding:**
- Categorical columns, including Gender, MaritalStatus, EmpJobRole, EducationBackground, Overtime, Employee Department, and Attrition, were converted into numerical values using various encoding techniques. These techniques include label encoding, one-hot encoding, and mapping encoding.

**Challenges Faced:**
- One significant challenge encountered during feature engineering was the treatment of the target variable, Performance Rating. It has three distinct values (e.g., [3, 4, 2]) that were transformed into [0, 1, 2] for use in a multiclass classification problem.

**Correlation Analysis:**
- Correlation analysis was conducted to understand the relationships between different features in the dataset. This helps in identifying how variables are related to one another and is essential for feature selection and model building.

**Heatmap Analysis:**
- The heatmap analysis was used to visualize the correlation between features. Features with a correlation coefficient above a certain threshold (0.7 in this case) were considered highly correlated. Identifying highly correlated features can be important for dimensionality reduction and model performance.

**Irrelevant Features:**
- In this dataset, no irrelevant features were identified. This means that all columns have some degree of relevance or potential predictive power for the target variable, and there was no need to drop any columns due to lack of relevance.

Overall, the feature engineering process involved preparing the data for machine learning, including encoding categorical variables and addressing the challenges presented by the target variable's multiple classes. Correlation and heatmap analyses were conducted to gain insights into the relationships between features, ensuring that the dataset is well-prepared for modeling.
